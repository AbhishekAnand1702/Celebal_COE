```python
taxi = sns.load_dataset('taxis')
taxi
```

```python
plt.figure(figsize=(16, 9), dpi=200)
plt.scatter(taxi['distance'], taxi['fare'], label='Fare')
plt.xlabel('Distance', size=20)
plt.ylabel('Fare', size=20)
plt.title('Distance vs fare', size=20)
plt.legend()
plt.show()
```

```python
taxi['color'].value_counts()
```

```python
plt.figure(figsize=(16, 9), dpi=200)
plt.scatter(taxi['distance'], taxi['fare'], label='Fare', c='r')
plt.xlabel('Distance', size=20)
plt.ylabel('Fare', size=20)
plt.title('Distance vs fare', size=20)
plt.legend()
plt.show()
```

```python
color = taxi['color'].map({
    'yellow':'y',
    'green':'g'
})
```

```python
plt.figure(figsize=(16, 9), dpi=200)
plt.scatter(taxi['distance'], taxi['fare'], label='Fare', c=color)
plt.xlabel('Distance', size=20)
plt.ylabel('Fare', size=20)
plt.title('Distance vs fare', size=20)
plt.legend()
plt.show()
```

# If we take random DataSet by using randint function and then visualizate the data

```python
jaipur_student = pd.DataFrame({
    'date':pd.date_range('010123', periods=10),
    'student':np.random.randint(40, 50, 10)
})

jaipur_student
```

```python
delhi_student = pd.DataFrame({
    'date':pd.date_range('010123', periods=10),
    'student':np.random.randint(30, 55, 10)
})

delhi_student
```

```python
plt.figure(figsize=(16, 9))
plt.bar(jaipur_student['date'], jaipur_student['student'])
plt.show()
```

```python
plt.figure(figsize=(16, 9), dpi=200)
plt.plot(jaipur_student['student'], label='jaipur')
plt.plot(delhi_student['student'], label='delhi')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 9), dpi=200)
plt.plot(jaipur_student['student'], label='jaipur')
plt.plot(delhi_student['student'], label='delhi')
plt.text(3, 44, 'jaipur', size=20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.bar(jaipur_student['date'], jaipur_student['student'])
xticks = plt.xticks()
plt.show()
```
